```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
resume_path = "data_set/Resume/Resume.csv"
df_resume = pd.read_csv(resume_path)
df_resume.head(3)
```

```python
df_resume.Category.unique()
```

```python
len(df_resume.Category.unique())
```

```python
category_counts = df_resume.Category.value_counts()

#Plotting the value counts using a bar plot
plt.figure(figsize=(12, 4))  
category_counts.plot(kind='bar')
plt.title('Category Value Counts')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=90) 
plt.show()
```
